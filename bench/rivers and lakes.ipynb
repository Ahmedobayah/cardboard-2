## Natural zoom levels of features

What's the "natural" zoom level of these features, not withstanding the constraint that they be contained within a tile? I'll compute these zoom levels using mercantile.

```python
import math
import mercantile

w, s, e, n = mercantile.bounds(0, 0, 0)
minx, miny = mercantile.xy(w, s)
maxx, maxy = mercantile.xy(e, n)
(maxx - minx), (maxy - miny)

def zoom_length(z):
    return (maxx-minx) / (2**z)

def length_zoom(l):
    """Return the natural zoom level of a feature of length `l`."""
    return int(math.log((maxx - minx) / l, 2))

print zoom_length(7)
print length_zoom(zoom_length(7))
```

```python
nat_zooms = [length_zoom(float(line.strip().split()[3])) for line in lines]
plt.hist(nat_zooms, normed=1)
plt.show()
```

## Tile index zoom levels

Now, reintroduce the constraint that features must be contained in one tile. How are the zoom levels of these tiles distributed?

```python
zooms = [int(line.strip().split()[1]) for line in lines]
plt.hist(zooms, normed=1)
plt.show()
```

## Natural Earth 1:10M river and lake centerlines

```python
def plot3(path, alpha=0.1):
    
    with open(path) as f:
        lines = f.readlines()
    
    nat_zooms = [length_zoom(float(line.strip().split()[3])) for line in lines]
    zooms = [int(line.strip().split()[1]) for line in lines]
    
    pylab.rcParams['figure.figsize'] = 8, 5

    plt.ylabel("frequency")
    plt.xlabel("zoom level")
    plt.hist(nat_zooms, bins=15, range=(0,15), normed=1)
    plt.title("a) Natural")
    plt.show()

    plt.ylabel("frequency")
    plt.xlabel("zoom level")
    plt.hist(zooms, bins=15, range=(0,15), normed=1)
    plt.title("a) Tile index")
    plt.show()

    plt.scatter(nat_zooms, zooms, alpha=alpha)
    plt.xlabel("Natural zoom level")
    plt.ylabel("Tile index zoom level")
    plt.title("c) Tile index vs Natural")
    plt.show()
```

```python
plot3('ne_10m_rivers_lake.txt', alpha=0.05)
```
